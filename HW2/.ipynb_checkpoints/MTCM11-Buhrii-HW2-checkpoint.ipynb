# 1. Для даних з вкладки “hedge_fund_series" порахувати такі статистики:
(а) математичне сподівання;

(b) волатильність;

(c) Коефіцієнт Шарпа. _Технічно коефіцієнт Шарпа рахується як відношення математичного сподівання до волатильності надлишкового прибутку. Тут ми використовуємо total returns, але умовно будемо називати цей mean-volatility ratio коефіцієнтом Шарпа._

Порахуйте ці статистики за період 1 рік (annualize).

```python
def basic_stats(data, annualized = False):
    # математичне сподівання
    data_mean = data.mean()

    # волатильність
    data_vol = data.std()

    # коефіцієнт Шарпа
    data_SR = data_mean/data_vol
    
    if annualized:
        data_mean = data_mean*12
        data_SR = data_SR*12
    
    return data_mean, data_vol, data_SR
```

```python
hfs_stats1 = pd.concat(basic_stats(hfs, annualized = True), axis = 1, keys = ["Annualized mean", "Volatility", "Sharpe ratio"])
hfs_stats1
```

# 2. Для даних з вкладки “hedge_fund_series" порахувати статистики, пов’язані з кінцевими ризиками (tail-risk):
(a) Асиметрія;

(b) Надлишковий ексцес (понад 3);

(c) VaR (0,05) –  5-ий квантиль прибутковості;

(d) CVaR (0,05) – середнє значення прибутків на рівні або нижче 5-го квантиля;

(e) Максимальна просадка – знайдіть дати максимального/мінімального/відновлення в межах періоду максимальної просадки. (maximum drawdown)

Ці статистики не потрібно перераховувати за рік.

```python
def statistics_stats(data):
    T = data.shape[0]
    
    data_mean = data.mean()
    data_vol = data.std()
    
    # асиметрія (Skewness)
    data_skewness = ((data - data_mean)**3).mean()/(data_vol**3)/(T-1) #pd.DataFrame.skew дає інші результати, бо метод підрахунку інший

    # ексцес (Kurtosis) та надлишковий ексцес (Excess kurtosis)
    data_kurtosis = ((data - data_mean)**4).mean()/(data_vol**4)/(T-1) #pd.DataFrame.kurtosis дає інші результати, бо метод підрахунку інший
    data_excess_kurtosis = data_kurtosis - 3

    return data_skewness, data_kurtosis, data_excess_kurtosis

hfs_stats2 = pd.concat(statistics_stats(hfs), axis = 1, keys = ["Skewness", "Kurtosis", "Excess kurtosis"])
hfs_stats2
```

```python
def return_stats(data):
    # значення ризику (Value at Risk (0,05))
    data_VaR = data.quantile(0.95)

    # очікуваний збиток (Expected Shortfall = Conditional Value at Risk (0,05))
    data_CVaR = data[data <= data_VaR].mean()

    # максимальна посадка (Maximum drawdown)
    data_max_drawdown = (data - data.cummax()).min()
    
    return data_VaR, data_CVaR, data_max_drawdown

hfs_stats3 = pd.concat(return_stats(hfs), axis = 1, keys = ["VaR", "CVaR", "Max drawdown"])
hfs_stats3
```

# 3. Для даних з вкладки “hedge_fund_series" реалізувати регресію кожного індекса з SPY (він знаходиться у вкладці "merrill_factors"). Додайте вільний член (intercept).

Порахувати наступні статистики на основі регресії:

(a) Market Beta

(b) Treynor Ratio

(c) Information ratio

Перерахуйте ці статистики за період 1 рік (annualize).

```python
# готуємо дані для регресії
targets = hfs.columns
factor = "SPY US Equity"
```

```python
# спершу переглянемо динаміку
plt.figure(figsize=(15, 8))
for target in targets:
    plt.plot(hfs.index, hfs[target], label = target)
plt.plot(mf.index, mf[factor], label = factor)
plt.legend()
plt.xlabel("Date")
plt.ylabel("Total return")
plt.show()
```

```python
def plot_results(x, y, alpha, beta, xlabel, ylabel, MSE):
    plt.figure(figsize = (12,5))
    plt.scatter(x, y, color = "blue")
    
    line_x = np.linspace(x.min(), x.max(), 100)
    plt.plot(line_x, alpha + line_x*beta, color = "green")

    plt.title("{} = {:.4f} + {:.4f}*{}; MSE = {:.4f}".format(ylabel, alpha, beta[0], xlabel, MSE))
    plt.xlabel(factor)
    plt.ylabel(target)
    plt.show()
    

def Simple_Linear_Regression(X, y, factor, target):
    X_train, X_test, y_train, y_test = train_test_split(X[[factor]], y[target], test_size=0.3, random_state=10)
    
    model = LinearRegression()
    model.fit(X_train, y_train)
    
    y_pred = model.predict(X_test)  
    test_MSE = mean_squared_error(y_test, y_pred)
    
    alpha = model.intercept_
    beta = model.coef_
    
    plot_results(X[factor], y[target], alpha, beta, factor, target, test_MSE)
    
    total_errors = y[target] - model.predict(X[[factor]])
    return alpha, beta, total_errors.std()

# результати регресії
coefficients = pd.DataFrame()
errors_std = pd.DataFrame()
for target in targets:
    a, b, e = Simple_Linear_Regression(mf, hfs, factor, target)
    coefficients.loc[target, "alpha"] = a
    coefficients.loc[target, factor] = b
    errors_std.loc[target, "error std"] = e
```

```python
# Market beta
coefficients
```

```python
# Treynor ratio
TR = pd.DataFrame(hfs[targets].mean()/coefficients[factor], columns = ["TR"])
TR
```

```python
# Information ratio
IR = pd.DataFrame(coefficients["alpha"] / errors_std["error std"], columns = ["IR"])
IR
```

# 6. Використовуючи лінійну регресію, відтворіть індекс HFRI за шістьома факторами, переліченими у вкладці "merrill_factors". Додайте вільний член:

$$r_t^{hrfi} = \alpha^{merr}+x_t^{merr}\beta^{merr}+\epsilon_t^{merr} \quad (1)$$

$$\hat{r}_t^{merr} = \hat{\alpha}^{merr}+x_t^{merr}\hat{\beta}^{merr}$$


Друге рівняння є лише позначенням для гіпотези лінійної регресії.

(a) Чому дорівнює вільний член та бета коефіцієнти?

(b) Чи реалістичні розміри позицій у бета-версіях чи вони вимагають величезних long і short позицій?

(c) Чому дорівнює R-squared?

(d) Обчисліть волатильність $\epsilon_t^{merr}$ (похибок).

```python
# готуємо дані для регресії
target = "HFRIFWI Index"
factors = mf.columns

X_train, X_test, y_train, y_test = train_test_split(mf[factors], hfs[target], test_size=0.3, random_state=10)
```

```python
# модель регресії
model1 = LinearRegression()
model1.fit(X_train, y_train)
```

(a) Коефіцієнти регресії

```python
def model_coef(model, with_alpha = True):
    if with_alpha:
        alpha = model.intercept_
    else:
        alpha = "-"
    beta = model.coef_

    coefficients = pd.DataFrame()
    coefficients.loc[target, "alpha"] = alpha
    for i in range(len(factors)):
        coefficients.loc[target, factors[i]] = beta[i]
    return coefficients
    
coefficients = model_coef(model1)
coefficients
```

(b) Зважаючи на величину коефіцієнтів можна сказати, що позиції реалістичні та не вимагають величезних long і short позицій.

(c) $R^2$

```python
y_pred = model1.predict(X_test)  
r2 = r2_score(y_test, y_pred)
r2
```

(d) волатильність похибок

```python
total_errors = hfs[target] - model1.predict(mf[factors])
error_std = total_errors.std()
error_std
```

# 7. Розглянемо авторегресію. Починаючи з $t = 61$ місяця вибірки, виконайте наступне:

A. Використайте дані за попередні 60 місяців, щоб натренувати лінійну регресію (1). Це дасть оцінки параметрів регресії, залежних від часу та $\tilde{\alpha}_t^{merr}$ та $\tilde{\beta}_t^{merr}$

B. Маючи ці параметри, а також залежну змінну $x_t^{merr}$ обчислити значення $\tilde{r}_t^{hfri} = \tilde{\alpha}^{merr} + (x_t^{merr})'\tilde{\beta}^{merr}$

C. Перейдіть до $t = 62$, але тепер використовуйте від $t = 2$ до $t = 61$ для оцінки. Зробіть те саме, що у пунктах A та B, і продовжуйте цей процес для усього ряду даних. Таким чином, ми запускаємо ковзаючу 60-місячну регресію для кожного моменту часу.

Наскільки добре працює реплікація поза вибіркою по відношенню до цільової змінної?

```python
predicted_r = []
for t in range(61, len(hfs)):
    X_train = mf.iloc[t-61:t, :]
    y_train = hfs.iloc[t-61:t, 0]
    
    model2 = LinearRegression()
    model2.fit(X_train, y_train)
    
    r_t = model2.predict(mf.iloc[t:t+1, :])
    predicted_r.append(r_t)
```

```python
plt.figure(figsize=(15, 8))
plt.plot(hfs.index[61:], predicted_r, label = "Predicted data")
plt.plot(hfs.index[61:], hfs.iloc[61:, 0], label = "Real data")
plt.legend()
plt.xlabel("Date")
plt.ylabel("Total return")
plt.show()
```

На графіку виведено порівняння передбачених прибутків (передбачення кожного наступного значення рахується на основі даних 60 минулих місяців). Як бачимо, передбачена динаміка є доволі схожою до справжньої.

```python
mean_squared_error(predicted_r, hfs.iloc[61:, 0])
```
